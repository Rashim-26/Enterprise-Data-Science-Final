```python
b= a.transpose()
b
```

```python
d=c.astype(int)
```

```python
e= d.loc[:,'Italy':].div(d.iloc[-1]['Italy':])*100
e
```

```python
final_plot= e.reset_index().rename(columns={'index':'date'})
final_plot
```

```python
fig = go.Figure()
## defines how to plot the individual trace
for each in country_list:
    fig.add_trace(go.Scatter(x=final_plot.date.iloc[:871],
                                y=final_plot[each],
                                mode='markers+lines',
                                opacity=0.9,
                                line_width=2,
                                marker_size=4, 
                                name=each
                                 )
                     )

## defines the overall layout properties
fig.update_layout(
    width=1024,
    height=900,
    xaxis_title="Date",
    yaxis_title="Percentage of Covid Cases (Absolute Cases/Population (%))",
    hovermode='x',
    
)
#fig.update_yaxes(type="log",range=[1.1,5.5])


fig.update_layout(xaxis_rangeslider_visible=True)
#fig.show(renderer='chrome')
fig.show()
```
